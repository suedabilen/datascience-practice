### Doğrusal İlişkilerin Gösterilmesi

```python
import seaborn as sns 
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips");
df = tips.copy()
```

```python
sns.lmplot(x = "total_bill" , y = "tip" ,data = df);
#lineer model kısaltmasıdır
```

```python
#açık renkli kısımlar sapmaları ifade ediyor
sns.lmplot(x = "total_bill" , y = "tip" ,data = df , hue = "smoker");
```

```python
sns.lmplot(x = "total_bill" , y = "tip" ,data = df ,col = "time" , hue = "smoker");
```

```python
sns.lmplot(x = "total_bill" , y = "tip" ,data = df ,col = "time" ,  row = "sex",hue = "smoker");
```

### HeatMap

```python
import seaborn as sns 
flights = sns.load_dataset('flights')
df = flights.copy()
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df["passengers"].describe().T
```

```python
#daha okunabilir hale getirmek
#pivot table aracılığıyla bunu yapalım.
df = df.pivot("month","year","passengers");
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df , annot = True , fmt ="d")
```

```python
sns.heatmap(df , annot = True , fmt ="d" , linewidths= .5);
```
